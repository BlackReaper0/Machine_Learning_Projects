## 🍀Outlier Detection and Handeling

```python
plt.figure(figsize=(20,10))
sns.boxplot(df)
```

```python
fig=px.box(df, x='Runtime',
             title="Original", orientation='h')
fig.update_layout(height=600)
fig.show()
```

```python
q1=df['Runtime'].quantile(0.25)
q3=df['Runtime'].quantile(0.75)
iqr=q3-q1
print("Q1:{} & Q2:{} & IQR:{}".format(q1,q3,iqr))

ul=q3+1.5*iqr
ll=q1-1.5*iqr
print("Up:{} & Low:{}".format(ul,ll))
```

```python
df_iqr_out=df.loc[(df['Runtime']>ul) | (df['Runtime']<ll)]
df_iqr_out
```

```python
df_iqr_out.shape
```

### Capping the Outliers

```python
df.loc[(df['Runtime']>ul), 'Runtime']=ul
df.loc[(df['Runtime']<ll), 'Runtime']=ll
```

```python
fig=px.box(df, x='Runtime',
             title="After Capping", orientation='h')
fig.update_layout(height=600)
fig.show()
```

## 🍀Building Model

```python
plt.figure(figsize=(20,15))
dataplot = sns.heatmap(df.corr(), annot=True, annot_kws={'size': 15}) 
  
# displaying heatmap 
plt.show()
```

```python
X=df.iloc[:, :-1].values
y=df.iloc[:, -1].values
```

```python
X, y
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error
model=XGBRegressor()
model.fit(X_train, y_train)
y_hat = model.predict(X_test)
rmse = mean_squared_error(y_test, y_hat, squared=False)
score=r2_score(y_test, y_hat)
```

```python
rmse
```

```python
score
```

## 🍀With Standardization

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

### 🌟XGBoostRegressor

```python
model_sc=XGBRegressor()
model_sc.fit(X_train, y_train)
y_pred_xgb = model_sc.predict(X_test)
xgb_rmse = mean_squared_error(y_test, y_pred_xgb, squared=False)
xgb_score=r2_score(y_test, y_pred_xgb)
```

```python
print('RMSE:{} & R2_Score:{}'.format(xgb_rmse, xgb_score))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_xgb, color='red', label=f'XGBoost Regressor (RMSE={xgb_rmse:.2f})')
plt.scatter(y_test, y_test, color='blue', label='Actual Values', alpha=0.7)
plt.plot(y_test, y_test, color='green', linestyle='--', alpha=0.5, label='Ideal Fit')

plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.legend()
```
